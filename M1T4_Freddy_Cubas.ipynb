```python
#ACTIVIDAD_4 Ejemplo de Ejercicios de operaciones y expresiones.

#MEDIA ARIMETCIA: Crea cinco variables con notas numéricas y calcula su media. 
# Ejemplo: “La nota media es 6.0”.

#Creamos las variables con las notas
nota1 = 60
nota2 = 76.8
nota3 = 95
nota4 = 100
nota5 = 55.4

# Sumanos las notas y dividimos entre 5 para obtener la media
media = (nota1 + nota2 + nota3 + nota4 + nota5) / 5

#Imprimimos el resultado
print("La nota media es", media)
```

```python
#ACTIVIDAD_5 REDONDEO:
# Redondea a dos decimales el resultado de: (365 / 12) * 14.7

#Definimos las variables, 
#Aunque lo podemos hacer directamente si  nesecidad de las variables
#lo hago de esta manera para mayor claridad.

Var1 = 365
Var2 = 12

resultado = (Var1 / Var2) * 14.7
print("Resultado redondeado:", round(resultado, 2))
#Imprimimos el resultado redondeado a dos decimales utilizando la función round

```

```python
#ACTIVIDAD_6: Operadores lógicos
#Crea variables `username` y `password` y verifica: Que username tenga entre 3 y 10 caracteres.
#Que password sea “Tokio” o “Python”.

#definimos las variables
username = "Freddy"
password = "Tokio"

#En este caso usamos condicionales para verificar las condiciones

# Primero decimos que si username es mayor o igual a 3 y menor o igual a 10 es valido (utilizmaos if <=)
#si es invalido (utilizamos else) y havvemos lo mismo con password.

if 3 <= len(username) <= 10:
    print("Username válido")
else:
    print("Username inválido")

if password in ["Tokio", "Python"]:
    print("Password válido")
else:
    print("Password inválido")
```
