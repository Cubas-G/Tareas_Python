```python
# EJERCICIO N_1: Creación de lista y tupla

frutas_lista = ["manzana", "pera", "uva"]
frutas_tupla = ("mango", "kiwi", "fresa")

print("Lista:", frutas_lista)
print("Tupla:", frutas_tupla)

# Mostrar elementos específicos

print("Segundo elemento de la lista:", frutas_lista[1])
print("Penúltimo elemento de la tupla:", frutas_tupla[-2])

# Modificar elementos (solo lista)
frutas_lista[0] = "sandía"
print("Lista modificada:", frutas_lista)

# Intento de modificar tupla (comentado porque da error)
# frutas_tupla[0] = "naranja"

# Tamaños
print("Tamaño de la lista:", len(frutas_lista))
print("Tamaño de la tupla:", len(frutas_tupla))

# Búsqueda
print("¿'pera' está en la lista?", "pera" in frutas_lista)
print("¿'kiwi' está en la tupla?", "kiwi" in frutas_tupla)

# Añadir elementos
frutas_lista.append("melón")
print("Lista con elemento añadido:", frutas_lista)

# Para añadir a una tupla, debemos crear una nueva
frutas_tupla = frutas_tupla + ("naranja",)
print("Tupla con elemento añadido:", frutas_tupla)

# Borrar contenido
frutas_lista.clear()
print("Lista vacía:", frutas_lista)

# Las tuplas no se pueden vaciar, pero podemos reasignar
frutas_tupla = ()
print("Tupla vacía:", frutas_tupla)

```

```python
# EJERCICIO N_2:Sets y Diccionarios
#Crea un set (≥ 3 strings) y un diccionario (≥ 3 pares clave:valor) y realiza

#Creación de set y diccionario

colores_set = {"rojo", "verde", "azul"}
persona_dic = {"nombre": "Carlos", "edad": 30, "ciudad": "Madrid"}

print("Set:", colores_set)
print("Diccionario:", persona_dic)

# Intento de mostrar segundo elemento del set (no se puede por no tener orden)
print("No se puede acceder por índice en un set.")

# Mostrar valor del primer par del diccionario
primer_valor = list(persona_dic.values())[0]
print("Primer valor del diccionario:", primer_valor)

# Modificar elementos
persona_dic["edad"] = 31
print("Diccionario modificado:", persona_dic)

# Tamaños
print("Tamaño del set:", len(colores_set))
print("Tamaño del diccionario:", len(persona_dic))

# Búsquedas
print("¿'rojo' está en el set?", "rojo" in colores_set)
print("¿'nombre' está en el diccionario?", "nombre" in persona_dic)
print("Valor de 'ciudad':", persona_dic.get("ciudad"))

# Añadir elementos
colores_set.add("amarillo")
persona_dic["profesion"] = "Ingeniero"

print("Set actualizado:", colores_set)
print("Diccionario actualizado:", persona_dic)

# Borrar contenido
colores_set.clear()
persona_dic.clear()

print("Set vacío:", colores_set)
print("Diccionario vacío:", persona_dic)
```

```python
# EJERCICIO N_3: Suma de eelemntos de una lista

# Pedir números al usuario
n1 = float(input("ingrese el primer número: "))
n2 = float(input("ingrese el segundo número: "))
n3 = float(input("ingrese el tercer número: "))

numeros = [n1, n2, n3]

sumatorio = sum(numeros)
print("La suma total es:", sumatorio)
```

```python
# EJERCICIO N_4: Media arítmetica de dinamica

# Pide al usuario tres números float, guárdalos en una lista (sin bucles)
#calcula la suma total. Guarda el resultado en sumatorio. Pista: usa sum(lista).

media = sumatorio / len(numeros)
print("La media aritmética es:", media)

```

```python
#EJERCICIO N_5: Lista de compras
#Crea una lista con al menos 6 productos: productos = ["pan", "leche", ...].

productos = ["pan", "leche", "huevos", "arroz", "pollo", "café"]

print("Segundo producto:", productos[1])
print("Último producto:", productos[-1])

# Modificar tercer producto
productos[2] = "jamón"
print("Lista actualizada:", productos)

# Intercambiar primero y último
productos[0], productos[-1] = productos[-1], productos[0]
print("Lista tras intercambio:", productos)

# Slicing
print("Tres primeros:", productos[:3])
print("Tres últimos:", productos[-3:])
print("Posiciones pares:", productos[::2])

```

```python
#Ejercicio N_6: Formateo de cadenas invertidas
#Supón que obtienes una cadena de texto corrupta (al revés). La cadena contiene nombre y año de nacimiento. Debes formatearla para obtener:
#Empleado Nombre Apellido nacido en el año Año

cadena = "zereP solraC, 5891"

# Invertir
cadena_correcta = cadena[::-1]

# Separar
nombre, año = cadena_correcta.split(", ")

print(f"Empleado {nombre} nacido en el año {año}")

```

```python
#EJERCICIO N_7: Operaciones con conjuntos

mamiferos = {"Gato", "León", "Caballo", "Ballena", "Tortuga"}
peces = {"Ballena", "Sardina"}

mamiferos.remove("Tortuga")
peces.add("Pez Payaso")

print("Mamíferos:", mamiferos)
print("Peces:", peces)

interseccion = mamiferos & peces
print("Elementos en ambos conjuntos:", interseccion)

```

```python
# EJERCICIO N_8: Comparacion de alugnos 
#Usamos intersección y diferencias entre conjuntos, y convertimos los resultados a listas.

python = {"Ana", "Luis", "Carlos", "Marta"}
ia = {"Luis", "Marta", "Sofía"}

ambos = python & ia
solo_python = python - ia
solo_ia = ia - python

print("Ambos cursos:", list(ambos))
print("Solo Python:", list(solo_python))
print("Solo IA:", list(solo_ia))

```

```python
# EJERCICIO N_10: Gestion de tareas Con un diccionario
#En este ejercicio, vamos a crear un gestor de tareas donde podremos añadir información progresivamente a un diccionario.

tareas = {
    "pendientes": [],
    "en_progreso": [],
    "completadas": []
}

nombre_tarea = input("Introduce el nombre de la tarea: ")
estado = input("Estado (pendientes / en_progreso / completadas): ")

if estado in tareas:
    tareas[estado].append(nombre_tarea)
else:
    print("Estado no válido.")

print("Diccionario actualizado:", tareas)
                              
```
